```python
df = df.dropna()
```

```python
df_messages = df.copy()
```

```python
df_messages.reset_index(inplace = True )
```

```python
corpus =[]
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
corpus = []
for i in range(len(df_messages)):
    review = re.sub('[^a-zA-Z]', ' ', df_messages["title"][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review  if word not in set(stopwords.words("english"))]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#creatubg count vectoirizer model
# model shoudl wordk well as it is a binary classification

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, ngram_range = (1,3))
X = cv.fit_transform(corpus).toarray()
```

```python
y = df_messages["label"]
y.shape
```

```python
#train test split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.20, random_state = 42)
 
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    See full source and example: 
    http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# this model can also be used for multi clasification
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
```

```python
from sklearn import metrics
import numpy as np
import itertools
import seaborn as sns 
```

```python
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
score  = metrics.accuracy_score(y_test, y_pred)
print("acuuracy score is {}".format(score))
cm = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True, xticklabels = ["fake", "real"], 
            yticklabels = ["fake", "real"], cmap="YlGnBu")
```

# Multinomial Classifier with Hyperparameter¶

```python
previous_score=0
for alpha in np.arange(0,1,0.1):
    sub_classifier=MultinomialNB(alpha=alpha)
    sub_classifier.fit(X_train,y_train)
    y_pred=sub_classifier.predict(X_test)
    score = metrics.accuracy_score(y_test, y_pred)
    if score>previous_score:
        classifier=sub_classifier
    print("Alpha: {}, Score : {}".format(alpha,score))
```
